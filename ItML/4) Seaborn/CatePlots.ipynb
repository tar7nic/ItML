```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x= 'sex', y='total_bill', data = tips, estimator= np.std)
```

```python
sns.countplot(x='sex', data= tips)
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data= tips, color= 'pink', hue='smoker',split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data= tips, hue='sex')
# jitter is the argument passed for the scattering of the dots 
```

```python
sns.swarmplot(x='day', y='total_bill', data = tips, hue = 'sex')
# dont use swarm plots for large datasets
```

```python
sns.violinplot(x='day',y='total_bill', data= tips, color='pink')
sns.swarmplot(x='day',y='total_bill', data= tips, color= 'black')
```
