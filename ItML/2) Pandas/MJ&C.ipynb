```python
left = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner', on='Key')
```

```python
l1 = pd.DataFrame({'Key1':['K0','K0','K1','K2'],
                    'Key2':['K0','K1','K0','K1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})

r1 = pd.DataFrame({'Key1':['K0','K1','K1','K2'],
                   'Key2':['K0','K0','K0','K0'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']})
```

```python
l1
```

```python
r1
```

```python
pd.merge(l1,r1,on=['Key1','Key2'])
```

```python
pd.merge(l1,r1,how='outer',on=['Key1','Key2'])
```

```python
pd.merge(l1,r1,how='right',on=['Key1','Key2'])
```
