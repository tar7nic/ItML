```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, color = 'green')
plt.xlabel('x')
plt.ylabel('y=x**2')
plt.title('x vs x**2')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color = 'blue')
plt.subplot(1,2,2)
plt.plot(y,x, color = 'red')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1,0.8,0.8])
axes.plot(x,y)
axes.set_ylabel('Y')
axes.set_xlabel('X')
axes.set_title('X vs Y')
```

```python
fig =  plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #[left bottom width height]
axes2 = fig.add_axes([0.45,0.17,0.4,0.4]) 

axes2.plot(x,y)
axes1.plot(y,x)

axes1.set_title('Lorge')
axes2.set_title('Smoll')

plt.tight_layout()
```

```python
# Figure size and DPI 
fig1 = plt.figure(figsize=(8,2)) 
ax = fig1.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig2 , axes = plt.subplots(nrows =2, ncols=1, figsize = (3,3))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig2.savefig('my_fig.png', dpi = 200)
```

```python
fig3 = plt.figure()

ax1 = fig3.add_axes([0,0,1,1])
ax1.plot(x,x**2,label='X vs Xsq', color = 'red')
ax1.plot(x,x**3, label = 'X vs Xcube', color = 'blue')

# ax1.legend(loc = 0)
ax1.legend(loc = (0.05,0.85))
```

```python
fig4 = plt.figure()
ax4 = fig4.add_axes([0,0,1,1])

ax4.plot(x,y, color = 'black', lw = 1, marker = 'o', markersize = 10, markerfacecolor = 'yellow',
         markeredgecolor = 'blue', markeredgewidth = 3)  #Alpha is used for transparency of the plotted graph

#ax4.set_xlim([0,20]) # limit on x axis
#ax4.set_ylim([0,5]) # limit on y axis
```
