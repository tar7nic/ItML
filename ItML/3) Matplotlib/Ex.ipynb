```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig2 = plt.figure()
ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig3 = plt.figure()
ax3 = fig3.add_axes([0,0,1,1])
ax4 = fig3.add_axes([0.2,0.5,0.4,0.4])

ax3.plot(x,z) 
ax4.plot(x,y)

ax4.set_xlim([20.0,22.0])
ax4.set_ylim([300,550])

ax3.set_xlim([0,100])
ax3.set_ylim([0,10000])
```

```python
t,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y, lw = 1, linestyle = '--', color = 'green')
axes[1].plot(x,z, lw = 2, marker = 'o', markersize = 10, markerfacecolor = 'red', markeredgecolor = 'black', color = 'black')

plt.tight_layout()

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('x  vs y')

axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('x  vs z')
```

```python
t,axes = plt.subplots(nrows=1, ncols=2, figsize = (9,2))
axes[0].plot(x,y, lw = 2, linestyle = '--', color = 'green')
axes[1].plot(x,z, lw = 2, marker = 'o', markersize = 3 , markerfacecolor = 'red', markeredgecolor = 'black', color = 'black')

plt.tight_layout()

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('x  vs y')

axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('x  vs z')
```
